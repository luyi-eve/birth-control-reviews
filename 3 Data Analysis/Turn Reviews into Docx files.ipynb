```python
import pandas as pd

c1_df = pd.read_csv("../2 Data Classification/OpenAI Classification/2 OpenAI Classifier/c1_negative_reviews.csv")
c1_df
```

```python

from docx import Document

# Create a new Word document
doc = Document()

# Add a title to the document
doc.add_heading('C1 All Negative Reviews', level=1)

# Iterate over the DataFrame rows
for index, row in c1_df[['app_name','title','review','date','userName']].iterrows():
    app_name = row['app_name']
    date = row['date']
    userName = row['userName']
    review_title = row['title']
    review = row['review']
    
    # Add app name as a heading in the Word document
    doc.add_heading(f'App Name: {app_name}', level=2)
    
    # Add the date text
    doc.add_paragraph(date)
    
    # Add the userName text
    doc.add_paragraph(userName)
    
    # Add the review title text
    doc.add_paragraph(review_title)
    
    # Add the review text
    doc.add_paragraph(review)
    
    # Optionally add a page break after each review
    doc.add_page_break()

# Save the document
#doc.save('c1_all_negative_reviews.docx')
```
