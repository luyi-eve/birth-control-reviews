```python
from docx import Document

# Create a new Word document
doc = Document()

# Add a title to the document
doc.add_heading('Reviews and App Names', level=1)

# Iterate over the DataFrame rows
for index, row in bc_overturn_df[['app_name', 'review']].iterrows():
    app_name = row['app_name']
    review = row['review']
    
    # Add app name as a heading in the Word document
    doc.add_heading(f'App Name: {app_name}', level=2)
    
    # Add the review text
    doc.add_paragraph(review)
    
    # Optionally add a page break after each review
    doc.add_page_break()

# Save the document
doc.save('App_Reviews.docx')
```
